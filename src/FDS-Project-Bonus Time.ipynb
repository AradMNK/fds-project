```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time


df = pd.DataFrame(columns=['title','descriptions','row_dict','info','url'])


def open_link(driver, url, wait=1):
    # Save the current window handle (main window)
    main_window = driver.current_window_handle

    # Open the link in a new tab
    driver.execute_script(f"window.open('{url}');")

    # Switch to the new tab (assuming it's the next one)
    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(wait)

    rows = driver.find_elements(By.CSS_SELECTOR, 'div.kt-unexpandable-row')
    row_dict = {}
    for row in rows:
        try:
            title = row.find_element(By.CSS_SELECTOR, 'p.kt-base-row__title').text
            value = row.find_element(By.CSS_SELECTOR, 'p.kt-unexpandable-row__value').text
            row_dict[title] = value
        except Exception as e:
            print("Skipping bad data")
        
        
    try:
        # Get the href attribute of the a element (link)
        info = driver.find_element(By.XPATH,
            '//*[@id="app"]/div[2]/div/main/article/div/div[1]/section[2]/div/div[2]/div/p').text
    except Exception as e:
        print("Skipping bad data")
        info = ''
    
    
    
    # Close the new tab
    driver.close()

    # Switch back to the main window
    driver.switch_to.window(main_window)
    
    return row_dict,info
    
def get_elements(driver, df, wait=1):
    # Find a elements that wrap article elements
    links = driver.find_elements(By.XPATH, '//*[@id="post-list-container-id"]/div[1]/div/div/div//a')

    for link in links:
        # Get the href attribute of the a element (link)
        try:
            # Get the href attribute of the a element (link)
            url = link.get_attribute('href')
        except Exception as e:
            print(f"Skipping a link due to error: {e}")
            continue

        # Find h2 element with class 'kt-post-card__title' within the link
        title = link.find_element(By.CSS_SELECTOR, 'h2.kt-post-card__title').text

        # Find div elements with class 'kt-post-card__description' within the link
        description_elements = link.find_elements(By.CSS_SELECTOR, 'div.kt-post-card__description')
        descriptions = [desc.text for desc in description_elements]

        # Print the title, descriptions, and link
        print(f'Title: {title}')
        #print('Descriptions:')
        #for desc in descriptions:
        #    print(desc)
        #print(f'Link: {url}')
        
        row_dict,info = open_link(driver,url)
        
        # Add a row to the DataFrame
        df.loc[len(df)] = [title, descriptions, row_dict, info, url]
        time.sleep(wait)


def extract_and_scroll(driver, df, N, wait):
    for i in range(N):
        print('_'*50)
        print(f'Iteration: {i+1}')
        # Find div elements
        get_elements(driver, df)

        # Extract and print elements
        #for element in div_elements:
        #    print(element.text)

        # Scroll down
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

        # Wait for page to load
        time.sleep(wait)

def main():
    # Set up Selenium WebDriver
    #driver = webdriver.Chrome('/chromedriver-win64/chromedriver.exe')
    driver = webdriver.Chrome()

    # Go to the webpage
    driver.get('https://divar.ir/s/tehran')

    # Extract elements and scroll N times
    extract_and_scroll(driver, df=df, N=10,wait=2)

    # Close the driver
    driver.quit()
    
    print

if __name__ == "__main__":
    main()
```
