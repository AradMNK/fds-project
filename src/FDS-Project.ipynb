**Calculate tweet counts for each company:** We can group by the tickers indicated by `$` and count the number of tweets.

```python
tweets['ticker'] = tweets['text'].str.findall(r'\$(\w+)')
tweets['ticker'] = tweets['ticker'].apply(' '.join)
tweets['ticker']
```

```python
tweets_split = tweets.assign(ticker=tweets['ticker'].str.split(' ')).explode('ticker')
tweets_split
```

```python
tweet_counts = tweets_split.groupby('ticker')['id'].count().reset_index().rename(columns={'id': 'tweet_count'})
```

**Find the most and least tweeted stocks:** We can use the `idxmax()` and `idxmin()` functions to find the most and least tweeted stocks.

```python
most_tweeted = tweet_counts.loc[tweet_counts['tweet_count'].idxmax()]
least_tweeted = tweet_counts.loc[tweet_counts['tweet_count'].idxmin()]
print(f'Most tweeted about:{most_tweeted}')
print(f'Least tweeted about:{least_tweeted}')
```

**Segment the companies based on tweet counts:** We can use pandas’ `cut` function to segment the companies into bins based on tweet counts.

```python
import numpy as np

bins = [0, 10, 100, 500, 1000, 5000, np.inf]
labels = ['<10', '10-100','100-500', '500-1000', '1000-5000', '>5000']
tweet_counts['segment'] = pd.cut(tweet_counts['tweet_count'], bins=bins, labels=labels)

tweet_counts_sorted = tweet_counts.sort_values('tweet_count')
indices = [-10,-100,-1000]
selected_stocks = tweet_counts_sorted.iloc[indices]
selected_stocks
```

**Visualize the results:** We can use matplotlib and seaborn to create bar plots of the most and least tweeted stocks, and a histogram of the segments.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a theme
sns.set_theme()

# Create a color palette
palette = sns.color_palette("husl", 8)

# Bar plot of most and least tweeted stocks
plt.figure(figsize=(12, 6))
sns.barplot(x='ticker', y='tweet_count', palette=palette,
            data=pd.concat([tweet_counts.loc[[most_tweeted.name, least_tweeted.name]], selected_stocks])
            .sort_values(ascending=False, by='tweet_count'))
plt.title('Most and Least Tweeted Stocks', fontsize=20)
plt.xlabel('Company Name', fontsize=15)
plt.ylabel('Tweet Count', fontsize=15)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Histogram of segments
plt.figure(figsize=(12, 6))
sns.histplot(tweet_counts['segment'], color='skyblue')
plt.title('Segmentation of Companies Based on Tweet Counts', fontsize=20)
plt.xlabel('Segments', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
# Select tweets for 'AAPL' and 'TSLA'
selected_tweets = tweets[tweets['ticker'].isin(['AAPL', 'TSLA'])]

selected_tweets['day_of_week'] = selected_tweets['datetime'].dt.dayofweek
selected_tweets['hour'] = selected_tweets['datetime'].dt.hour

# Map numeric day of week to actual day name
day_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
selected_tweets['day_of_week'] = selected_tweets['day_of_week'].map(day_map)
```

```python
# Group by day of week and hour to get tweet frequencies
tweet_freq = selected_tweets.groupby(['day_of_week', 'hour']).size().reset_index(name='freq')
```

```python
# Pivot the data for the heatmap
tweet_freq_pivot = tweet_freq.pivot('day_of_week', 'hour', 'freq')
# Reorder the index to start from Monday
tweet_freq_pivot = tweet_freq_pivot.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(tweet_freq_pivot, cmap='rocket')

plt.title('Tweet Frequencies for AAPL and TSLA')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

```python
# Plot number of tweets per day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x='day_of_week',y='freq', data=tweet_freq, palette='mako',
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Number of Tweets per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Tweets')
plt.show()

# Plot number of tweets per hour of the day
plt.figure(figsize=(10, 6))
sns.barplot(x='hour',y='freq', data=tweet_freq, palette='mako',)
plt.title('Number of Tweets per Hour of the Day')
# Highlight the trading hours (9:30 AM to 4:00 PM ET corresponds to 14:30 to 21:00 UTC)
plt.axvspan(14.5, 21, color='green', alpha=0.2)
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Tweets')
plt.show()
```

*Co-occurence of various stocks in the same tweets.*

```python
from collections import Counter
from itertools import combinations

# Split the 'ticker' column into a list of tickers
tweets['tickers'] = tweets['ticker'].str.split()

# Get the top N most common tickers
N = 50
top_N_tickers = [ticker for ticker, _ in Counter(ticker for tickers in tweets['tickers'] for ticker in tickers).most_common(N)]

# Filter the tweets to include only the top N tickers
tweets['tickers'] = tweets['tickers'].apply(lambda tickers: [ticker for ticker in tickers if ticker in top_N_tickers])

# Initialize a Counter for each pair of tickers
co_occurrence = Counter()

# Count the co-occurrences
for tickers in tweets['tickers']:
    # Get all combinations of tickers in the current tweet
    ticker_combinations = combinations(tickers, 2)
    
    # Update the co-occurrence counter
    co_occurrence.update(ticker_combinations)
```

```python
co_occurrence
```

```python
# Convert the co-occurrence Counter to a DataFrame
co_occurrence_df = pd.DataFrame.from_records(list(co_occurrence.items()), columns=['tickers', 'freq'])
co_occurrence_df
```

```python
# Split the 'tickers' column into two separate columns
co_occurrence_df[['ticker1', 'ticker2']] = pd.DataFrame(co_occurrence_df['tickers'].tolist(), index=co_occurrence_df.index)
co_occurrence_df
```

```python
# Pivot the DataFrame to create a co-occurrence matrix
co_occurrence_matrix = co_occurrence_df.pivot('ticker1', 'ticker2', 'freq')
co_occurrence_matrix
```

```python
# Calculate the total co-occurrence frequency for each ticker
total_co_occurrence = co_occurrence_matrix.sum(axis=0) + co_occurrence_matrix.sum(axis=1)

# Sort the tickers by total co-occurrence frequency
sorted_tickers = total_co_occurrence.sort_values(ascending=False).index

# Reorder the rows and columns of the co-occurrence matrix
co_occurrence_matrix = co_occurrence_matrix.loc[sorted_tickers, sorted_tickers]
```

```python
# Plot the co-occurrence matrix using a heatmap
plt.figure(figsize=(20, 20)) 
sns.heatmap(co_occurrence_matrix, cmap='Blues', cbar_kws={"shrink": .5}, annot_kws={"size": 8})
plt.title('Co-occurrence of Stocks in Tweets', fontsize=14)
plt.xlabel('Ticker', fontsize=12)
plt.ylabel('Ticker', fontsize=12)
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```
