**Calculate tweet counts for each company:** We can group by the tickers indicated by `$` and count the number of tweets.

```python
tweets['ticker'] = tweets['text'].str.findall(r'\$(\w+)')
tweets['ticker'] = tweets['ticker'].apply(' '.join)
tweets['ticker']
```

```python
tweets_split = tweets.assign(ticker=tweets['ticker'].str.split(' ')).explode('ticker')
tweets_split
```

```python
tweet_counts = tweets_split.groupby('ticker')['id'].count().reset_index().rename(columns={'id': 'tweet_count'})
```

**Find the most and least tweeted stocks:** We can use the `idxmax()` and `idxmin()` functions to find the most and least tweeted stocks.

```python
most_tweeted = tweet_counts.loc[tweet_counts['tweet_count'].idxmax()]
least_tweeted = tweet_counts.loc[tweet_counts['tweet_count'].idxmin()]
print(f'Most tweeted about:{most_tweeted}')
print(f'Least tweeted about:{least_tweeted}')
```

**Segment the companies based on tweet counts:** We can use pandas’ `cut` function to segment the companies into bins based on tweet counts.

```python
import numpy as np

bins = [0, 10, 100, 500, 1000, 5000, np.inf]
labels = ['<10', '10-100','100-500', '500-1000', '1000-5000', '>5000']
tweet_counts['segment'] = pd.cut(tweet_counts['tweet_count'], bins=bins, labels=labels)

tweet_counts_sorted = tweet_counts.sort_values('tweet_count')
indices = [-10,-100,-1000]
selected_stocks = tweet_counts_sorted.iloc[indices]
selected_stocks
```

**Visualize the results:** We can use matplotlib and seaborn to create bar plots of the most and least tweeted stocks, and a histogram of the segments.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a theme
sns.set_theme()

# Create a color palette
palette = sns.color_palette("husl", 8)

# Bar plot of most and least tweeted stocks
plt.figure(figsize=(12, 6))
sns.barplot(x='ticker', y='tweet_count', palette=palette,
            data=pd.concat([tweet_counts.loc[[most_tweeted.name, least_tweeted.name]], selected_stocks])
            .sort_values(ascending=False, by='tweet_count'))
plt.title('Most and Least Tweeted Stocks', fontsize=20)
plt.xlabel('Company Name', fontsize=15)
plt.ylabel('Tweet Count', fontsize=15)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Histogram of segments
plt.figure(figsize=(12, 6))
sns.histplot(tweet_counts['segment'], color='skyblue')
plt.title('Segmentation of Companies Based on Tweet Counts', fontsize=20)
plt.xlabel('Segments', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

*Statistics on distributions of 5 individual stocks over time. Choose the individual stocks to perform reflect different sectors of the economy.*

We will manually select 5 stocks that represent different sectors of the economy. choose Apple (Technology), ExxonMobil (Energy), Pfizer (Healthcare), JPMorgan Chase (Financial), and Ford (Automotive).

```python
selected_stocks = ['AAPL', 'XOM', 'PFE', 'JPM', 'F']
df_selected = tweets[tweets['ticker'].isin(selected_stocks)]
```

```python
df_selected['created_at'] = pd.to_datetime(df_selected['created_at'])
df_selected['day'] = df_selected['created_at'].dt.to_period('D')

# Convert to datetime
df_selected['day'] = df_selected['day'].dt.to_timestamp()
tweet_counts = df_selected.groupby(['ticker', 'day'])['id'].count().reset_index().rename(columns={'id': 'tweet_count'})
```

```python
tweet_counts
```

```python
plt.figure(figsize=(15, 8))
sns.lineplot(x='day', y='tweet_count', hue='ticker', data=tweet_counts)
plt.title('Tweet Counts of Selected Stocks Over Time', fontsize=20)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Tweet Count', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Cashtag', title_fontsize='13', fontsize='12')
plt.show()
```

*Statitistics on distributions of all financial tweets over time.*

For this one I will take on a 7-day moving average approach, just to show that I know how to do that.

```python
tweet_counts['7_day_avg'] = tweet_counts['tweet_count'].rolling(window=7).mean()

# Set a theme
sns.set_theme()

# Create a color palette
palette = sns.color_palette("husl", 8)

# Line plot of 7-day moving average
plt.figure(figsize=(15, 8))
sns.lineplot(x='day', y='7_day_avg', data=tweet_counts, color='skyblue', linewidth=2.5)
plt.title('7-Day Moving Average of Financial Tweets Over Time', fontsize=20)
plt.xlabel('Date', fontsize=15)
plt.ylabel('7-Day Average Tweet Count', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Highlight the peak of the 7-day average tweet count
max_7_day_avg = tweet_counts['7_day_avg'].max()
max_date = tweet_counts[tweet_counts['7_day_avg'] == max_7_day_avg]['day'].values[0]
plt.scatter(max_date, max_7_day_avg, color='red')
plt.text(max_date, max_7_day_avg, 'Peak', verticalalignment='bottom', horizontalalignment='right', color='red', fontsize=15)

plt.show()
```
